### Okt 형태소 분석기 사용

```python
from konlpy.tag import Okt
okt= Okt() # 형태소, 명사만, 어절, 품사태깅
# okt= Okt(stem= True) 어간 추출
```

```python
data= '오늘 날씨가 추워서 기분이 좋지않다.'
```

```python
# 형태소 단위로 분석 - 형태소: 의미를 가진 가장 작은 단위
okt.morphs(data)
```

```python
# 명사 추출
okt.nouns(data)
```

```python
# 어절 추출
okt.phrases(data)
```

```python
# 품사태깅
okt.pos(data)
```

### 단어 빈도 계산

```python
from collections import Counter
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('ratings.txt', delimiter= '\t')
df
```

```python
x= df['document'][:3000]
```

```python
n= []

for i in x:
    n.append(i)
```

```python
okt= Okt()
noun= okt.nouns(str(n))
```

```python
count= Counter(noun)
```

```python
# 명사 빈도 카운트
noun_list= count.most_common(100)

for i in noun_list:
    print(i)
```

### 단어 리스트 형성

```python
word_list= []
word_list= df['document'][:3000]
```

```python
t= []

for i in word_list:
    morph= okt.pos(i)
    t.append(morph)
```

### 명사 추출

```python
noun_list= []
for s in t:
    for word, tag in s:
        if tag in ['Noun']:
            noun_list.append(word)
noun_list
```

### 두 글자 이상 단어만 추출

```python
noun_list2= [n for n in noun_list if len(n) >1]
noun_list2
```

### 단어별 개수 세기

```python
counts= Counter(noun_list2)
counts
```

```python
tags= counts.most_common(200)
tags
```

```python
dic= dict(tags)
dic
```

### 워드 클라우드 생성

```python
from PIL import Image # 워드 클라우드용 마스크를 생성하기 위한 라이브러리
import matplotlib.pyplot as plt

custom_mask= np.array(Image.open('도베르만.jpg'))
plt.figure(figsize= (8, 8))
plt.imshow(custom_mask, cmap= plt.cm.gray, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```

```python
from wordcloud import ImageColorGenerator
image_colors= ImageColorGenerator(custom_mask)
# 마스크용 사진의 색을 입력
```

```python
from wordcloud import WordCloud

wordcloud= WordCloud(font_path= 'NanumSquareNeo-aLt.ttf',
                     background_color= 'white',
                     width= 800, height= 600,
                     max_words= 200,
                     mask= custom_mask,
                     max_font_size= 30)
print(dic)

cloud= wordcloud.generate_from_frequencies(dic)
plt.figure(figsize= (12, 12))
plt.axis('off')
plt.imshow(cloud.recolor(color_func= image_colors), interpolation= 'bilinear')
plt.show()
```
