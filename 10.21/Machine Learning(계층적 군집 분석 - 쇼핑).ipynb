```python
import mglearn
import graphviz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_moons, make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
plt.rcParams['font.family']= 'Malgun Gothic'
plt.rcParams['axes.unicode_minus']= False
```

```python
customer_data= pd.read_csv('shopping-data.csv')
customer_data.head()
```

```python
customer_data.shape
```

```python
# 데이터 전처리
data= customer_data.iloc[:, 3:5].values
```

```python
# 덴드로그램을 사용하여 쇼핑 데이터 시각화
plt.figure(figsize= (10, 7))
plt.title('Customer shopping data dendrogram')
d= dendrogram(linkage(data, method= 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# 거리를 구하기 위해 유클리디안(euclidean) 사용
cluster= AgglomerativeClustering(n_clusters= 5,
                                 affinity= 'euclidean',
                                 linkage= 'ward')
```

```python
cluster.fit_predict(data)
```

```python
#시각화
plt.figure(figsize= (10,7))
plt.scatter(data[:, 0], data[:, 1], c= cluster.labels_, cmap= 'rainbow')
plt.show()
```

### 보스턴 데이터

```python
from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings('ignore', category= FutureWarning)
```

```python
# 1. 데이터 불러오기
# 2. 학습 데이터 프레임화
# 3. 학습데이터의 정답값
boston= load_boston()
data= pd.DataFrame(boston['data'], columns= boston['feature_names'])
target= pd.DataFrame(boston['target'], columns= ['price'])
```

```python
# 덴드로그램 사용하여 시각화 (method= 'ward')
plt.figure(figsize= (10, 7))
plt.title('Boston data dendrogram')
d= dendrogram(linkage(data, method= 'ward'))
```

```python
fig= plt.figure(figsize= (20,5))

ax1= fig.add_subplot(1, 3, 1)
d1= linkage(data, method= 'ward')
dr1= dendrogram(d1, orientation= 'left')
# ax1.set_yticklabels(target[dr1['leaves']], minor= False)

ax2= fig.add_subplot(1, 3, 2)
d2= linkage(data, method= 'ward')
dr2= dendrogram(d2, orientation= 'right')
# ax2.set_yticklabels(target[dr2['leaves']], minor= False)

ax3= fig.add_subplot(1, 3, 3)
d3= linkage(data, method= 'ward')
dr3= dendrogram(d3, orientation= 'bottom')
```

```python
# 클러스터 생생
cluster= AgglomerativeClustering(n_clusters= 5,
                                 affinity= 'euclidean',
                                 linkage= 'ward')
```

```python
# 생성한 클러스터에 데이터 학습 및 예측값 적용
cluster.fit_predict(data)
```
