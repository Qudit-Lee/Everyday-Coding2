### 1. 결측치 처리
- 제거
- 채우기

```python
import pandas as pd
import numpy as np
```

```python
d ={'score1': [100,90,np.nan,95],
    'score2': [30,np.nan,45,56],
    'score3': [52,40,80,98],
    'score4': [np.nan,np.nan,np.nan,65]}
```

```python
df= pd.DataFrame(d)
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.score1.isnull()]
```

#### 행 기준으로 삭제

```python
df.dropna()
```

```python
df.dropna(axis= 0, how= 'any')
```

#### 열 기준으로 삭제

```python
df.dropna(axis= 1)
```

#### 행의 전체 값이 결측치인 행을 삭제

```python
d2= {'score1': [100, np.nan, np.nan, 95],
     'score2': [np.nan, np.nan, np.nan, np.nan],
     'score3': [52, np.nan, 80, 98],
     'score4': [np.nan, np.nan, np.nan, 65]}
```

```python
df2= pd.DataFrame(d2)
df2
```

```python
df2.dropna(how= 'all')
```

#### 임계치 설정해서 제거

```python
df2.dropna(thresh= 2) # 적당한 임계값이 얼마인지는 분석가의 정확한 판단이 필요
```

#### 특정 열 안에서의 삭제

```python
df.dropna(subset= ['score2', 'score4'])
```

### 결측치 채우기

#### 특정한 단일 값으로 채우기

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna(method= 'pad')
```

```python
df.fillna(method= 'bfill')
```

#### 결측치를 각 열의 평균값으로 채우기

```python
df.fillna(df.mean())
```

#### 결측치를 중간값, 최소값, 최대값

```python
df.fillna(df.median())
```

```python
df.fillna(df.min())
```

```python
df.fillna(df.max())
```

#### replace() 함수로 결측치 채우기

```python
df.replace(to_replace= np.nan, value= 10)
```

#### interpolate() 함수로 결측치 채우기
- 선형 방법을 사용하여 결측값 채움
- 선형 방식은 인덱스를 무시하고 결측값들을 같은 간격으로 처리

```python
df.interpolate(method= 'linear', limit_direction= 'forward') 
```

```python
df.interpolate(method= 'pad')
```

### 2. 범주 특성을 OneHotEncoder로 변환
- 모든 데이터를 0과 1로 변환
- 컴퓨터는 모든 데이터를 이진법을 사용하여 처리하기 때문에 레이블 인코딩 보다 원핫 인코딩을 더 많이 사용

```python
df= pd.read_csv('mushrooms.csv')
```

```python
df
```

```python
one= pd.get_dummies(df)
```

```python
one
```

pandas.get_dummies의 문제점
- train 데이터엔 있지만 test 데이터에는 없는 카테고리를 원핫 인코딩된 컬럼으로 변환이 안됨

#### sklearn OneHotEncoder 사용

```python
x= df.iloc[:, 1:]
y= df['class']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#one= OneHotEncoder()
one= OneHotEncoder(sparse=False)
train_cat= one.fit_transform(x[['cap-shape']])
```

```python
train_cat
```

```python
one.categories_
```

```python
o= pd.DataFrame(train_cat, columns= ['cap-shape_'+ col for col in one.categories_[0]])
```

```python
one_x= pd.concat([x.drop(columns= ['cap-shape']), o], axis= 1)
```

```python
one_x
```

#### 범주특성 레이블 인코딩으로 변환
- 숫자의 크기가 크고 작은 것에 대한 특성이 적용
- 회귀와 같이 연속된 실수를 다루는 알고리즘에서 숫자 크기에 따른 순서나 중요도로 인식 될 수 있어서 잘못된 결과가 나올 수 있음.

```python
one_x['cap-surface'].unique()
```

```python
one_x['cap-surface'].value_counts()
```

```python
one_x['cap-surface']
```

```python
one_x['cap-surface'] = one_x['cap-surface'].map({'y': 0,'s':1,'f':2,'g':3})
```

```python
one_x
```

#### sklearn LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder= LabelEncoder()
encoder.fit(one_x['cap-color'])
one_x['cap-color']= encoder.transform(one_x['cap-color'])
```

```python
one_x
```
