```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings('ignore', category= FutureWarning)
```

```python
boston= load_boston()
```

```python
boston.keys()
```

```python
boston['target']
```

```python
boston['feature_names']
```

### 데이터 전처리 - 스킵

### 탐색적 분석

```python
x= pd.DataFrame(boston['data'], columns= boston['feature_names'])
x
```

```python
y=  pd.DataFrame(boston['target'], columns= ['price'])
y
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y,
                                                   test_size= 0.2,
                                                   random_state= 0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### 모델 생성

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_ # 가중치
```

```python
lr.intercept_ # 편향값(절편)
```

### 모델 평가

```python
train= lr.predict(x_train)
test= lr.predict(x_test)
```

```python
train_score= mean_squared_error(train, y_train)
test_score= mean_squared_error(test, y_test)
```

```python
print(train_score) # MSE
print(test_score)
```

```python
print('rmse: ', train_score**0.5) # RMSE
print('      ', test_score**0.5)
print('오차: ', abs(train_score**0.5-test_score**0.5))
```

```python
print(lr.score(x_train, y_train)) # R2 score
print(lr.score(x_test, y_test))
```

### 선형회귀 모델 특성 확장
- 선형회귀는 특성이 적으면 다른 알고리즘에 비해 성능이 낮게 나오며, 특성 확장을 통해 모델의 복잡도를 증가시켜 성능 향상 시켜야 함
- 각 컬럼들의 데이터를 곱하여 새로운 특성으로 확장

```python
extended_x_train= x_train.copy() # copy함수를 사용하여 변경된 값이 이중으로 적용되는 것을 방지
```

```python
extended_x_train.head()
```

```python
# extended_x_train의 각 컬럼들을 서로 한 번씩 곱해서 새로운 컬럼 추가

for col1 in x_train.columns: # column이 13개이므로 13번 반복
    for col2 in x_train.columns:
        extended_x_train[col1 +'x'+ col2]= x_train[col1]* x_train[col2]
```

```python
extended_x_train.head()
```

```python
pd.set_option('display.max_columns', None) # ...으로 숨겨진 값 보기
extended_x_train.head()
```

```python
extended_x_train.shape
```

### 특성 확장된 Train 데이터의 값과 정답으로 모델 학습

```python
lr2= LinearRegression().fit(extended_x_train, y_train)
```

```python
extended_x_test= x_test.copy()
```

```python
for col1 in x_train.columns: # column이 13개이므로 13번 반복
    for col2 in x_train.columns:
        extended_x_test[col1 +'x'+ col2]= x_test[col1]* x_test[col2]
```

```python
extended_x_test.shape
```

### 오차= 예측 - 정답

```python
ex_pre_train= lr2.predict(extended_x_train)
ex_pre_test= lr2.predict(extended_x_test)
```

```python
ex_train_mse= mean_squared_error(ex_pre_train, y_train)
ex_test_mse= mean_squared_error(ex_pre_test, y_test)
```

```python
print('train mse: ', ex_train_mse)
print('test mse: ', ex_test_mse)
print('train, test mse의 차이: ', abs(ex_train_mse-ex_test_mse))
```

```python
print('train rmse: ', ex_train_mse**0.5) 
print('test rmse: ', ex_test_mse**0.5)
print('train, test rmse의 차이: ', abs(ex_train_mse**0.5-ex_test_mse**0.5))
```

```python
print(lr2.score(extended_x_train, y_train))
print(lr2.score(extended_x_test, y_test))
```

#### 훈련 데이터의 값이 높으니 과대적합의 가능성이 높음.

### Lasso(L1) 규제

```python
from sklearn.linear_model import Lasso
```

```python
lasso= Lasso(alpha= 10) # alpha 값이 높아지면 과대적합이 줄어듬
```

```python
lasso.fit(extended_x_train, y_train)
```

```python
print('사용한 특성의 수: ', np.sum(lasso.coef_ != 0)) # 가중치가 0이 아닌 것
```

```python
pre_train_lasso= lasso.predict(extended_x_train) # 예측값
pre_test_lasso= lasso.predict(extended_x_test)
```

```python
lasso_train_mse= mean_squared_error(pre_train_lasso, y_train) # MSE
lasso_test_mse= mean_squared_error(pre_test_lasso, y_test)
```

```python
print('train mse: ', lasso_train_mse) 
print('test mse: ', lasso_test_mse)
print('train, test mse의 차이: ', abs(lasso_train_mse-lasso_test_mse))
```

```python
print('train rmse: ', lasso_train_mse**0.5) # RMSE
print('test rmse: ', lasso_test_mse**0.5)
print('train, test rmse의 차이: ', abs(lasso_train_mse**0.5-lasso_test_mse**0.5))
```

```python
print(lasso.score(extended_x_train, y_train)) # R2 score
print(lasso.score(extended_x_test, y_test))
```

```python
print('train rmse: ', ridge_train_mse**0.5)
print('test rmse: ', ridge_test_mse**0.5)
print('train, test rmse의 차이: ', abs(ridge_train_mse**0.5-ridge_test_mse**0.5))
```

```python
print(ridge.score(extended_x_train, y_train))
print(ridge.score(extended_x_test, y_test))
```

위 2개의 규제가 동일한 alpha 값을 가지고 있을 때 L1(Lasso)규제에서 과대적합이 더 잘 해소됨.

### 하이퍼 파라미터 튜닝
- alpha 값을 바꿔가면서 rmse 확인

Lasso(랏쏘)

```python
alpha_list= [0.001, 0.01, 0.1, 1, 10, 100, 1000]
lasso_train_list= []
lasso_test_list= []
```

```python
for i in alpha_list:
    lasso= Lasso(alpha= i)
    lasso.fit(extended_x_train, y_train)
    
    # train 데이터
    train_pred= lasso.predict(extended_x_train)
    lasso_train_rmse= mean_squared_error(train_pred, y_train)**0.5
    lasso_train_list.append(lasso_train_rmse)
    
    # test 데이터
    test_pred= lasso.predict(extended_x_test)
    lasso_test_rmse= mean_squared_error(test_pred, y_test)**0.5
    lasso_test_list.append(lasso_test_rmse)
```

```python
print(lasso_train_list)
print(lasso_test_list)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
```

```python
plt.figure(figsize= (10,5))
plt.plot(lasso_train_list, label='train')
plt.plot(lasso_test_list, label='test')
plt.title('Lasso 모델 alpha - RMSE 그래프')
plt.xlabel('alpha 값')
plt.xticks(np.arange(7), alpha_list) # x축 범위 설정(x측 값의 개수와 바꿔줄 값)
plt.ylabel('RMSE')
plt.grid()
plt.show()
```

Ridge(릿지)

```python
alpha_list= [0.001, 0.01, 0.1, 1, 10, 100, 1000]
ridge_train_list= []
ridge_test_list= []
```

```python
for i in alpha_list:
    ridge= Ridge(alpha= i)
    ridge.fit(extended_x_train, y_train)
    
    # train 데이터
    train_pred= ridge.predict(extended_x_train)
    ridge_train_rmse= mean_squared_error(train_pred, y_train)**0.5
    ridge_train_list.append(ridge_train_rmse)
    
    # test 데이터
    test_pred= ridge.predict(extended_x_test)
    ridge_test_rmse= mean_squared_error(test_pred, y_test)**0.5
    ridge_test_list.append(ridge_test_rmse)
```

```python
print(ridge_train_list)
print(ridge_test_list)
```

```python
plt.figure(figsize= (10,5))
plt.plot(ridge_train_list, label='train')
plt.plot(ridge_test_list, label='test')
plt.title('Ridge 모델 alpha - RMSE 그래프')
plt.xlabel('alpha 값')
plt.xticks(np.arange(7), alpha_list)
plt.ylabel('RMSE')
plt.grid()
plt.show()
```
